```python
datos_mileage_mean = datos['Mileage'].mean()
datos['Mileage'].map(lambda p: p - datos_mileage_mean)
```

```python
datos_mileage_median = datos['Mileage'].median()
datos['Mileage'].map(lambda p: p - datos_mileage_median)
```

```python
def remean_mileage(row):
  row['Mileage'] = row['Mileage'] - datos_mileage_mean
  return row

datos.apply(remean_mileage, axis='columns')
```
