```python
import spacy
```

```python
!python -m spacy download de_core_news_sm
```

```python
nlp = spacy.load("de_core_news_sm")
```

```python
document = "Fernbusse sind sehr populär in Europa. Diese Busse fahren zu günstigen Preisen von einer Stadt in eine andere Stadt. In Deutschland benutzen wir vor allem den Flixbus. Du kannst Tickets online, oder mit der App kaufen. Für lange Distanzen ist der Zug eine gute Option. Züge sind schnell und bequem. Tickets kannst du online bei der Deutschen Bahn kaufen. Wenn du früh buchst, dann findest du tolle Angebote."
```

```python
doc = nlp(document)
```

```python
for i,s in enumerate(doc.sents):
  print(i,s)
```

```python
sentences = list(doc.sents)

for i in range (1, 6):
  if i < len(sentences):
    sentence = sentences[i]
    print("Tokens de la frase", {i+1})
    for token in sentence:
      print(token.text, token.pos_)
```

```python
from spacy.matcher import Matcher
```

```python
matcher = Matcher(nlp.vocab)
```

```python
pattern = [{"POS": "DET"}, {"POS": "NOUN"}, {"OP": "?"}]
```

```python
matcher.add("CADENA", [pattern])
```

```python
matches = matcher(doc)
```

```python
print("Objetos encontrados: ", len(matches))
```

```python
for match_id, start, end in matches:
  print("Texto de los objetos: ", doc[start:end].text)
```
